```python
import numpy

# Define the directory and file names
dataDir = 'C:/Users/navne/Python Files/Data Acquisition BIA 627/Final Project/'
file_name_test = dataDir + 'testTrack_hierarchy.txt'
file_name_train = dataDir + 'trainIdx2_matrix.txt'
output_file = dataDir + 'output_p.txt'

# Open files for reading and writing
fTest = open(file_name_test, 'r')
fTrain = open(file_name_train, 'r')
fOut = open(output_file, 'w')

# Initialize variables
trackID_vec = [0] * 6
albumID_vec = [0] * 6
artistID_vec = [0] * 6
lastUserID = -1
user_rating_inTrain = numpy.zeros(shape=(6, 3))

# Read and process the test file
for line in fTest:
    arr_test = line.strip().split('|')
    userID, trackID, albumID, artistID = arr_test[0], arr_test[1], arr_test[2], arr_test[3]
    
    if userID != lastUserID:
        ii = 0  # Reset counter for a new user
        user_rating_inTrain = numpy.zeros(shape=(6, 3))  # Reset user ratings

    # Update track, album, and artist vectors
    trackID_vec[ii] = trackID
    albumID_vec[ii] = albumID
    artistID_vec[ii] = artistID
    ii += 1
    lastUserID = userID  # Update last seen userID

    if ii == 6:
        Trainline = fTrain.readline()
        while Trainline:
            arr_train = Trainline.strip().split('|')
            trainUserID, trainItemID, trainRating = arr_train[0], arr_train[1], arr_train[2]
            
            if trainUserID < userID:
                Trainline = fTrain.readline()
                continue

            if trainUserID == userID:
                for nn in range(6):
                    if trainItemID == trackID_vec[nn]:
                        user_rating_inTrain[nn, 0] = trainRating
                    if trainItemID == artistID_vec[nn]:
                        user_rating_inTrain[nn, 1] = trainRating

            if trainUserID > userID:
                for nn in range(6):
                    outStr = f"{userID}|{trackID_vec[nn]}|{user_rating_inTrain[nn, 0]}|{user_rating_inTrain[nn, 1]}"
                    fOut.write(outStr + '\n')
                break  # Break the while loop to process the next user in the test file

            Trainline = fTrain.readline()  # Read the next line in the train file

# Close all opened files
fTest.close()
fTrain.close()
fOut.close()
```
